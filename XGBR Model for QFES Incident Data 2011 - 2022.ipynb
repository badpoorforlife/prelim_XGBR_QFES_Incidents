```python
# using model on default settings is resulting in overfitting
# Start by tuning "gamma", "max_depth" and min_child_weight to try and prevent overfitting
# if that fails, then try tuning for "subsample", "colsample_bytree" and "eta"(with num_round increase) to introduce noise
```

```python
# Use GridSearchCV to tune hyperparameters of the model
from xgboost import XGBRegressor
from sklearn.model_selection import GridSearchCV

# Define the hyperparameters to search
param_grid = {'n_estimators': [30], # define number of estimators to use in forest (default = 100)
              'gamma': [0], # define 'gamma', the minimum loss reduction for next leaf node on tree (default = 0)
              'max_depth': [6], # define maximum depth of trees (default = 6)
              'min_child_weight': [0.01], # define minimum child weight (default = 1)
              'subsample': [1], # define subsample (default = 1), e.g. if 0.5, model will only use half of dataset
              'colsample_bytree': [1], # define column sample by tree
              'learning_rate': [0.3], # define learning rate
              'alpha': [11000], # L1 regularization term on weights (default = 0)
              'lambda': [1], # L2 regularization term on weights (default = 1)
              'tree_method': ["gpu_hist"], # define tree_method as 'gpu_hist' to utilise GPU
              'enable_categorical': [True], # enable the use of categorical variables
              'random_state': [77]}  # use random_state to set seed for reproducibility

# Use GridSearchCV to perform hyperparameter search and k-fold cross-validation
grid_search = GridSearchCV(XGBRegressor(), param_grid, cv=3, scoring='neg_root_mean_squared_error', verbose=3)
grid_search.fit(X, y)

# Get the best hyperparameters
best_params = grid_search.best_params_
print(best_params)
```
